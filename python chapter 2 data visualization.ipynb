```python
# python basic data visualization 
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Line plot
plt.plot(x, y, marker='o', color='blue', linestyle='-', label='Data')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot Example')
plt.legend()
plt.show()

```

```python
# Scatter plot
plt.scatter(x, y, color='red', marker='s', label='Data')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Scatter Plot Example')
plt.legend()
plt.show()

```

```python
# Bar chart
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
values = [30, 50, 80, 40, 60]
plt.bar(categories, values, color='green', alpha=0.7)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()

```

```python
# Pie chart
labels = ['Label 1', 'Label 2', 'Label 3', 'Label 4', 'Label 5']
sizes = [20, 30, 15, 10, 25]
colors = ['red', 'blue', 'green', 'yellow', 'purple']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
```
