# subplots  methods and iteration and indexing

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
```

```python
x = np.linspace(0,10,20)
y= randint(1,50,20)
y = np.sort(y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1]) # left, buttom , width , height
axes.plot(x,y)
axes.set_xlabel('x axis')
axes.set_ylabel('Y axis')
axes.set_title('x vs Y axis')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=3)
```

```python
# to avoid overlapping
fig,axes = plt.subplots(nrows=2,ncols=3)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(x,y)
axes[0].set_title('Plot1')
axes[1].set_title('Plot2')
```
