```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from numpy.random import randint
```

```python
x = np.linspace(0,10,20)
```

```python
x
```

```python
y = randint(0,50,20)
```

```python
y
```

```python
y = np.sort(y)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('x vs Y axis')
plt.show()
```

```python
plt.plot(x,y,color='m')
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('x vs Y axis')
plt.show()
```

```python
plt.plot(x,y,color='m',linestyle='--')
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('x vs Y axis')
plt.show()
```

```python
plt.plot(x,y,color='m',linestyle='--', marker='*')
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('x vs Y axis')
plt.show()
```

```python
plt.plot(x,y,color='m',linestyle='--', marker='*', markersize=10)
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('x vs Y axis')
plt.show()
```

```python
plt.plot(x,y,color='m',linestyle='--', marker='*', markersize=10, linewidth=1.5)
plt.xlabel('x axis')
plt.ylabel('Y axis')
plt.title('x vs Y axis')
plt.show()
```

# multiple plots on save canvas

```python
plt.subplot(1,2,1) #number or rows, number of columns,plot we are refering
plt.plot(x,y,color='r')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color='r')
plt.subplot(1,2,2)
plt.plot(x,y,color='m')
```
