# Scatter plots

```python
import matplotlib.pyplot as plt
```

```python
y_views = [534,693,402,555,660]
f_views = [445,345,234,678,980]
t_views = [123,340,55,555,678]
days =[1,2,3,4,5]
```

```python
plt.scatter(days,y_views,label ='Youtubele views',marker='o')
```

```python
plt.scatter(days,y_views,label ='Youtubele views',marker='o')
plt.scatter(days,f_views,label ='facebook views',marker='o')
```

```python
plt.scatter(days,y_views,label ='Youtubele views',marker='o')
plt.scatter(days,f_views,label ='Facebook views',marker='o')
plt.scatter(days,t_views,label ='twitter views',marker='o')
plt.xlabel('Days')
plt.ylabel('views')
plt.title('social media views')
plt.legend()
```

```python
plt.scatter(days,y_views,label ='Youtubele views',marker='o')
plt.scatter(days,f_views,label ='Facebook views',marker='o')
plt.scatter(days,t_views,label ='twitter views',marker='o')
plt.xlabel('Days')
plt.ylabel('views')
plt.title('social media views')
plt.grid(color ='r',linestyle='--')
plt.legend()
```

#  histogram 

```python
points = [22,33,44,23,43,56,78,34,67]
bins =[0,20,40,60,80,100,120]
```

```python
plt.hist(points,bins)
plt.xlabel('bins')
plt.ylabel('Frequency')
plt.title('bins vs frequency')
```

```python
plt.hist(points,bins)
plt.xlabel('bins')
plt.ylabel('Frequency')
plt.title('bins vs frequency')
plt.show() #it wiss only show grap not other unnecessary details strings
```

# pie chart

```python
labels_1 = ['facebook','instagram','youtube','linked']
views = [300,350,400,500]
```

```python
plt.pie(views,labels= labels_1)
```

```python
plt.pie(views,labels= labels_1)
plt.show()
```

```python
labels_1 = ['facebook','instagram','youtube','linked']
views = [300,350,400,500]
explode_1 = [0,0,0,0.2]
plt.pie(views,labels= labels_1,autopct ='%1.1f%%')
plt.show()
```

```python
labels_1 = ['facebook','instagram','youtube','linked']
views = [300,350,400,500]
explode = [0,0,0,0.2]
plt.pie(views,labels= labels_1,autopct ='%1.1f%%',explode =explode_1 )
plt.show()
```

```python
labels_1 = ['facebook','instagram','youtube','linked']
views = [300,350,400,500]
explode = [0,0,0,0.2]
plt.pie(views,labels= labels_1,autopct ='%1.1f%%',explode =explode_1,shadow=True )
plt.show()
```
