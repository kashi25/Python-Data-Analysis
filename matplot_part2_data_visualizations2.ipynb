```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
```

```python
x = np.linspace(0,10,20)
y= randint(1,50,20)
y = np.sort(y)
```

```python
x
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1]) # left, buttom , width , height
axes.plot(x,y)
axes.set_xlabel('x axis')
axes.set_ylabel('Y axis')
axes.set_title('x vs Y axis')
```

```python
# two sets for one canvas
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3]) #left from biger canvas, buttom height from bigger canvas, width, height
```

```python
# two sets for one canvas
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])
ax1.plot(x,y,color='r')
ax1.set_xlabel('X axis')
ax1.set_ylabel('y axis')
ax1.set_title('plot 1')



ax2.plot(x,y,color='m')
ax2.set_xlabel('X axis')
ax2.set_ylabel('y axis')
ax2.set_title('plot 2')
```
