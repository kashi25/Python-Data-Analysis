# figure size and aspect ratio and DPI

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randint
```

```python
x = np.linspace(0,10,20)
y = randint(1,50,20)
y = np.sort(y)
```

```python
fig = plt.figure(figsize=(3,2),dpi=100) #figsize(width,height),dpi is dots per inch
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure(figsize=(5,3),dpi=100) #figsize(width,height),dpi is dots per inch
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# to save the fig
fig.savefig('big_pic.png',dpi=100)
```

```python
# with subplots

fig,axes = plt.subplots(nrows=1,ncols=2,figsize =(7,2))
axes[0].plot(x,y)
axes[1].plot(x,y)
```

```python
# to save the fig
fig.savefig('my_pic.png',dpi=100)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_title('x vs y')
```

```python
# legend , It helps to show graph more clearly
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X vs x square')
ax.plot(x,x**3,label='X vs x cube')
ax.legend()
```
